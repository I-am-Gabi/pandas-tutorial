#Usando pandas com a base de dados Movielens

Que tal usarmos o pandas de forma mais aplicada agora, vamos responder algumas perguntas sobre a base de dados do Movielens. Lembrando que nós já lemos o nossos dados em um DataFrame e fizemos um merge.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('max_columns', 50)
```

```python
# pass in column names for each CSV
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('ml-100k/u.user', sep='|', names=u_cols)

r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols)

# the movies file contains columns indicating the movie's genres
# let's only load the first five columns of the file with usecols
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
movies = pd.read_csv('ml-100k/u.item', sep='|', usecols=range(5))
movies.columns = m_cols

# create one merged DataFrame
movie_ratings = pd.merge(movies, ratings)
lens = pd.merge(movie_ratings, users)
```

## Quais são os 25 filmes mais avaliados ?

```python
most_rated = lens.groupby('title').size().order(ascending=False)[:25]
print most_rated
```

Muita coisa aconteceu no código acima, mas nada muito estranho. Primeiramente, estamos quebrando o DataFrame em grupos pelo título do filme (split) e depois aplicando o método size (apply) para obter a total de registros de cada grupo. Daí nós ordenamos os resultados de forma decrescente e apresentamos apenas os 25 primeiros usando o slicing de Python.

Em SQL, isto seria equivalente ao:

    SELECT title, count(1)
    FROM lens
    GROUP BY title
    ORDER BY 2 DESC
    LIMIT 25;

pandas também tem outros métodos para isto como o value_counts() que também realiza a mesma operação.

```python
lens.title.value_counts()[:25]
```

##Quais são os filmes com maiores notas ?

```python
movie_stats = lens.groupby('title').agg({'rating': [np.size, np.mean]})
print movie_stats.head()
```

Nós usamos o método agg para passar um dicionário especificando as colunas a serem agregadas (como chaves) e uma lista de funções que gostaríamos de aplicar.

Vamos ordenar o DataFrame resultante para que possamos visualizar os filmes com as maiores notas.

```python
# sort by rating average
print movie_stats.sort([('rating', 'mean')], ascending=False).head()
```

Já que movie_stats é um DataFrame, nós usamos o método sort - apenas objetos Series podem usar a função order. Outra informação útil é que como nossas colunas agora são MultiIndex, nós precisamos passá-los em uma tupla especificando como ordená-los.

Os filmes acima são avaliados  tão raramente, que nem podemos qualificá-los como top filmes. Que tal apenas olharmos os filmes que foram avaliados pelo menos 100 vezes.

```python
atleast_100 = movie_stats['rating'].size >= 100
print movie_stats[atleast_100].sort([('rating', 'mean')], ascending=False)[:15]
```

Agora sim! Os resultados parecem mais reais. Veja que nós usamos o boolean indexing para filtrar o nosso frame movie_stats.

Nós quebramos a pergunta em várias partes, e vimos que para Python conseguir os top 15 filmes com maiores avaliações médias com pelo menos 100 avaliações foi somente:


    movie_stats = lens.groupby('title').agg({'rating', [np.size, np.mean})

    atleast_100 = movie_stats['rating'].size >= 100

    movie_stats[atleast_100].sort([('rating', 'mean')], ascending=False)[:15]
    
Olha como fica o SQL:

    SELECT title, COUNT(1) size, AVG(rating) mean
    FROM lens
    GROUP BY title
    HAVING COUNT(1) >= 100
    ORDER BY 3 DESC
    LIMIT 15;

##Limitando nossa população de seguir em diante

Vamos brincar com os nossos dados. Vamos olhar apenas para os 50 filmes mais avaliados. Em cima deles vamos construir uma Series de filmes que atendem um limiar que possamos usar para filtrar alguns filmes.

```python
most_50 = lens.groupby('movie_id').size().order(ascending=False)[:50]
```

O nosso SQL aqui equivalente seria:
    CREATE TABLE most_50 AS (
        SELECT movie_id, COUNT(1)
        FROM lens
        GROUP BY movie_id
        ORDER BY 2 DESC
        LIMIT 50
    );

Esta tabela permite que nós possamos usar EXISTS, IN ou JOIN como quisermos para filtrar nossos resultados. Veja um exemplo usando EXISTS:

    SELECT *
    FROM lens
    WHERE EXISTS (SELECT 1 FROM most_50 WHERE lens.movie_id = most_50.movie_id);

###Binarizando nossos usuários

Eu não acho muito legal comparar nossas idades individuais, portanto vamos binarizar nossos usuários em groupos por idade usando o pandas.cut.

```python
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
lens['age_group'] = pd.cut(lens.age, range(0, 81, 10), right=False, labels=labels)
print lens[['age', 'age_group']].drop_duplicates()[:10]
```

pandas.cut permite que você binarize seus dados numéricos. Nas linhas acima, primeiramente nós criamos rótulos para os nossos bins. Depois nós quebramos nosos usuários em 8 grupos de 10 anos (0-9, 10-29, 20-29, etc.). O right=False indica que queremos que os grupos sejam exclusivos no limite da idade máxima de cada grupo, isto é (20,29] (logo um usuário de 30 anos cai no grupo dos 30.

Agora vamos comparar as avaliações sobre os grupos de idades.

```python
print lens.groupby('age_group').agg({'rating': [np.size, np.mean]})
```

Jovens usuários tendem ser mais críticos que os outros grupos por faixa etária.Vamos olhar apenas para os 50 filmes mais avaliados e ver como eles são vistos por cada grupo. Lembrando que temos o most_50 (Series) que criamos antes para filtro.


```python
lens.set_index('movie_id', inplace=True)
```

```python
by_age = lens.ix[most_50.index].groupby(['title', 'age_group'])
by_age.rating.mean().head(15)
```

Note que tanto títulos e grupos de idade são os índices aqui, onde o valor médio de notas é uma Series. Isto iria produzir uma grande lista se olhassemos a base inteira.

Não seria legal olhar os nossos dados como uma tabela ? Cada título como uma linha, cada grupo de faixa etária como uma coluna e cada nota média em uma célula.

Conheça o método unstack!

```python
by_age.rating.mean().unstack(1).fillna(0)[10:20]
```

O que o unstack faz ?

unstack (desempilhar em português) realiza uma operação de retirar uma camada específica do MultiIndex ( por padrão, o groupby faz com que o campo pelo qual foi agrupado se torne um index - visto que nós agrupamos por 2 campos title e age_group, ele se tornou um MultiIndex). Nós desempilhamos o segundo índice (lembre-se que Python utiliza índices a partir de 0) e depois nós preenchemos os valores NULL com 0.

Se nós tivessemos feito:

    by_age.rating.mean().unstack(0).fillna(0)

Nós teríamos nossos grupos de idade como linhas e títulos do filme como colunas.

##Quais são os filmes que mulheres e homens mais se diferem em gostos ?

Imagine como você faria isto com SQL. Você precisaria de uma série de comandos IF/CASE com funções de agregação para pivotar seu banco de dados. Sua query seria algo assim: 

    SELECT title, AVG(IF(sex = 'F', rating, NULL)), AVG(IF(sex = 'M', rating, NULL))
    FROM lens
    GROUP BY title;
    
Imagine ainda como seria fazer isto com mais de 2 colunas. Aí que entra o pivot_table em DataFrame que permite que estes tipos de operações sejam bem mais fáceis e menos verborrágicos!

```python
lens.reset_index('movie_id', inplace=True)
```

```python
pivoted = lens.pivot_table(rows=['movie_id', 'title'],
                           cols=['sex'],
                           values='rating',
                           fill_value=0)
print pivoted.head()
```

```python
pivoted['diff'] = pivoted.M - pivoted.F
print pivoted.head()
```

```python
pivoted.reset_index('movie_id', inplace=True)
```

```python
disagreements = pivoted[pivoted.movie_id.isin(most_50.index)]['diff']
disagreements.order().plot(kind='barh', figsize=[9, 15])
plt.title('Male vs. Female Avg. Ratings\n(Difference > 0 = Favored by Men)')
plt.ylabel('Title')
plt.xlabel('Average Rating Difference');
```

É claro que os homens gostam mais de Exterminador do Futuro que as mulheres. Mas Independence Day as mulheres gostarem mais ? kkkk essa base era de mulheres Geekies só pode!
