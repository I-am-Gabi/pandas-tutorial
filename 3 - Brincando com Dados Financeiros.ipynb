##Operações Financeiras

Vamos trabalhar com algumas operações financeiras comuns para exemplificar o uso do pandas. Uma delas é a média flutuante (moving average).

```python
close_px = df['Adj Close']
```

```python
mavg = pd.rolling_mean(close_px, 40)
mavg[-10:]
```

```python
rets = close_px / close_px.shift(1) - 1
rets.head()
```

```python
close_px.pct_change().head()
```

```python
close_px.plot(label='AAPL')
mavg.plot(label='mavg')
plt.legend()
```

```python
df = pd.io.data.get_data_yahoo(['AAPL', 'GE', 'GOOG', 'IBM', 'KO', 'MSFT', 'PEP'], 
                               start=datetime.datetime(2010, 1, 1), 
                               end=datetime.datetime(2013, 1, 1))['Adj Close']
df.head()
```

```python
rets = df.pct_change()
```

```python
plt.scatter(rets.PEP, rets.KO)
plt.xlabel('Returns PEP')
plt.ylabel('Returns KO')
```

```python
pd.scatter_matrix(rets, diagonal='kde', figsize=(10, 10));
```

```python
corr = rets.corr()
corr
```

```python
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns);
```

```python
plt.scatter(rets.mean(), rets.std())
plt.xlabel('Expected returns')
plt.ylabel('Risk')
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (20, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```
